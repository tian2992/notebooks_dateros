```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
pengu = sns.load_dataset("penguins")
pengu
pengu.dropna(inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
pengu['species'] = lb.fit_transform(pengu["species"])
```

```python
peng_cols =  ['bill_length_mm', 'bill_depth_mm','flipper_length_mm']  # ["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"]
peng = pengu[peng_cols]
peng_target = pd.DataFrame(pengu["species"])

# peng.reset_index(inplace=True)
# peng_target.reset_index(inplace=True)
```

```python

from sklearn.decomposition import PCA


pca_ = PCA(n_components=2)
principal_penguin = pca_.fit_transform(peng)
principal_penguin_df = pd.DataFrame(principal_penguin, columns=['comp1', 'comp2'])
```

Procedemos a ahora graficar este dataset pero con la reduccion de variables a 2 dimensiones usando PCA


```python
plt.figure(figsize=(10,10))
plt.xlabel("Componente dimensional 1")
plt.ylabel("Componente dimensional 2")
plt.title("Analisis de componentes de pinguinos 2D")


targets = [0,1,2]
colors = ['r','g','purple']
for target, color in zip(targets, colors):
    indexTK = (peng_target == target) ##indexTK puede ser valor 0, 1 u 2, esto nos da el el rango de indices por cada uno
    plt.scatter(principal_penguin_df.loc[indexTK, "comp1"],
                principal_penguin_df.loc[indexTK, "comp2"],
                c = color, s = 50)
```
